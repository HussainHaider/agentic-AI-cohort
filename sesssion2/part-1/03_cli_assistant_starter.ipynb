## Setup

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

print("✅ Setup complete!")
```

## Helper Functions

```python
from openai import RateLimitError, APIError

def make_api_call(messages, temperature=0.7, max_tokens=150):
    """Makes API call with error handling"""
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages,
            temperature=temperature,
            max_tokens=max_tokens
        )
        return response.choices[0].message.content
    except RateLimitError:
        return "⚠️ Rate limit exceeded."
    except Exception as e:
        return f"⚠️ Error: {e}"

def print_header(text):
    """Print nice header"""
    print("\n" + "="*60)
    print(f"  {text}")
    print("="*60 + "\n")

print("✅ Helper functions ready!")
```

## Mode 1: Chat

```python
def chat_mode():
    """Chat with conversation memory"""
    print_header("CHAT MODE")
    print("Start chatting! Type 'quit' to exit.\n")
    
    # TODO: Create conversation with system prompt
    conversation = [
        {"role": "system", "content": "You are helpful."}
    ]
    
    while True:
        user_input = input("You: ").strip()
        
        if user_input.lower() in ['quit', 'exit', 'q']:
            print("\nGoodbye! 👋\n")
            break
        
        if not user_input:
            continue
        
        # TODO: Add user message
        conversation.append({"role": "user", "content": user_input})
        
        # TODO: Call API
        response = make_api_call(conversation)
        
        print(f"\nAI: {response}\n")
        
        # TODO: Add AI response
        conversation.append({"role": "assistant", "content": response})

# Uncomment to test:
# chat_mode()
```

## Mode 2: Summarize

```python
def summarize_mode():
    """Summarize text"""
    print_header("SUMMARIZE MODE")
    
    # Get text
    print("Paste text (press Enter, type END, press Enter):")
    lines = []
    while True:
        line = input()
        if line == "END":
            break
        lines.append(line)
    
    text = "\n".join(lines)
    
    if not text.strip():
        print("⚠️ No text!\n")
        return
    
    print("\n⏳ Summarizing...\n")
    
    # TODO: Create system prompt
    system = "You create concise summaries."
    
    # TODO: Call API
    messages = [
        {"role": "system", "content": system},
        {"role": "user", "content": f"Summarize:\n{text}"}
    ]
    summary = make_api_call(messages, temperature=0.3, max_tokens=200)
    
    print("SUMMARY:")
    print("="*60)
    print(summary)
    print("="*60 + "\n")

# Uncomment to test:
# summarize_mode()
```

## Mode 3: Email Writer

```python
def email_mode():
    """Generate professional emails"""
    print_header("EMAIL WRITER")
    
    description = input("What email do you need? ").strip()
    
    if not description:
        print("⚠️ Describe the email!\n")
        return
    
    print("\n⏳ Writing...\n")
    
    # TODO: System prompt for professional emails
    system = """You write professional emails with:
    - Subject line
    - Professional greeting
    - Clear body
    - Proper closing"""
    
    # TODO: Call API
    messages = [
        {"role": "system", "content": system},
        {"role": "user", "content": f"Write email for: {description}"}
    ]
    email = make_api_call(messages, max_tokens=300)
    
    print("EMAIL:")
    print("="*60)
    print(email)
    print("="*60 + "\n")

# Uncomment to test:
# email_mode()
```

## Main Menu

```python
def main():
    """Main menu"""
    print_header("CLI ASSISTANT")
    
    while True:
        print("\nChoose mode:")
        print("1. Chat")
        print("2. Summarize")
        print("3. Email Writer")
        print("4. Quit")
        
        choice = input("\nChoice (1-4): ").strip()
        
        if choice == '1':
            chat_mode()
        elif choice == '2':
            summarize_mode()
        elif choice == '3':
            email_mode()
        elif choice == '4':
            print("\nGoodbye! 👋\n")
            break
        else:
            print("⚠️ Invalid choice.")

# Run it!
# main()
```
