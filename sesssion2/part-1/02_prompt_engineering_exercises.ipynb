## Setup

```python
from openai import OpenAI, RateLimitError, APIError
import os
from dotenv import load_dotenv
import json

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def call_gpt(messages, temperature=0.7, max_tokens=150):
    """Helper function for API calls"""
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages,
            temperature=temperature,
            max_tokens=max_tokens
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {e}"

print("✅ Ready to practice!")
```

## Exercise 1: Basic vs Improved Prompts

```python
# Bad prompt
bad = "Tell me about AI"
response = call_gpt([{"role": "user", "content": bad}])
print("BAD PROMPT:", bad)
print("Response:", response)
print("\n" + "="*70 + "\n")
```

```python
# Good prompt
good = """Explain AI in exactly 3 sentences for someone who's never heard of it. 
Use simple language and include one real-world example."""
response = call_gpt([{"role": "user", "content": good}])
print("GOOD PROMPT:", good)
print("Response:", response)
print("\n💡 Notice the difference!")
```

## Exercise 2: Few-Shot Learning

```python
# Sentiment classification example
prompt = """Classify sentiment:

Text: 'I love this!' -> positive
Text: 'This is terrible.' -> negative
Text: 'It's okay.' -> neutral
Text: 'Best ever!' -> ?"""

response = call_gpt([{"role": "user", "content": prompt}], max_tokens=10)
print(response)
```

## Exercise 3: Chain of Thought

```python
# Without chain of thought
simple = "What is 15% of 85?"
response = call_gpt([{"role": "user", "content": simple}])
print("WITHOUT:", response)
print("\n" + "="*70 + "\n")
```

```python
# With chain of thought
cot = "What is 15% of 85? Let's solve step by step:"
response = call_gpt([{"role": "user", "content": cot}], max_tokens=100)
print("WITH:", response)
```

## Exercise 4: System Prompts

```python
# Professional email writer
system = """You are a professional email writer. 
Write clear, concise emails with proper structure."""

user = "Decline a meeting politely due to conflict."

response = call_gpt([
    {"role": "system", "content": system},
    {"role": "user", "content": user}
], max_tokens=200)

print(response)
```

## Exercise 5: JSON Output

```python
# Movie info in JSON
prompt = """Give info about 'Inception' in JSON:
{"title": "", "year": 0, "director": "", "genre": [], "rating": 0}
Return ONLY valid JSON."""

response = call_gpt([
    {"role": "system", "content": "Respond only in JSON."},
    {"role": "user", "content": prompt}
], max_tokens=150)

print("Raw:", response)

try:
    data = json.loads(response)
    print(f"\nTitle: {data['title']}")
    print(f"Year: {data['year']}")
    print("✅ Parsed successfully!")
except:
    print("❌ Parse failed")
```
