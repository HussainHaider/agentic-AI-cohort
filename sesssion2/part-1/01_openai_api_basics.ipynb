```python
from openai import OpenAI
import os
from dotenv import load_dotenv
import json

# Load API key from .env file
load_dotenv()

# Create OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Verify it loaded
if client.api_key:
    print("✅ API key loaded successfully!")
    print(f"Key starts with: {client.api_key[:8]}...")
else:
    print("❌ API key not found! Check your .env file.")
```

## 1. Your First API Call

```python
# The simplest possible API call
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "user", "content": "Hello! Introduce yourself in one sentence."}
    ]
)

print("AI says:")
print(response.choices[0].message.content)
```

## 2. Understanding the Response

```python
# Let's examine the full response
print("Full response structure:")
print(response)

print("\n" + "="*50)
print("Key parts:")
print(f"Message: {response.choices[0].message.content}")
print(f"Tokens used: {response.usage.total_tokens}")
print(f"Model: {response.model}")
```

## 3. System Prompts

```python
# Example 1: Helpful assistant
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Tell me about AI."}
    ]
)
print("Helpful Assistant:")
print(response.choices[0].message.content)
```

```python
# Example 2: Pirate personality
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a pirate. Speak like a pirate and say 'Arrr!' often."},
        {"role": "user", "content": "Tell me about AI."}
    ]
)
print("\nPirate Assistant:")
print(response.choices[0].message.content)
```

## 4. Temperature Control

```python
prompt = "Write a one-sentence description of Paris."

# Low temperature - consistent
print("Low Temperature (0.1) - Run 3 times:")
for i in range(3):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1
    )
    print(f"{i+1}. {response.choices[0].message.content}")
```

```python
# High temperature - creative
print("\nHigh Temperature (1.8) - Run 3 times:")
for i in range(3):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=1.8
    )
    print(f"{i+1}. {response.choices[0].message.content}")
```

## 5. Max Tokens

```python
# Short response
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": "Explain machine learning."}],
    max_tokens=50
)
print("With max_tokens=50:")
print(response.choices[0].message.content)
print(f"Tokens: {response.usage.completion_tokens}")
```

## 6. Conversation History

```python
# Start conversation
conversation = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "My name is Hammad."}
]

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=conversation
)

print("Assistant:", response.choices[0].message.content)
conversation.append({"role": "assistant", "content": response.choices[0].message.content})
```

```python
# Continue conversation
conversation.append({"role": "user", "content": "What's my name?"})

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=conversation
)

print("\nAssistant:", response.choices[0].message.content)
print("\n✅ The AI remembered!")
```

## 7. Error Handling

```python
from openai import RateLimitError, APIError, AuthenticationError

def safe_api_call(messages, max_tokens=100):
    """API call with error handling"""
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages,
            max_tokens=max_tokens
        )
        return response.choices[0].message.content
    except RateLimitError:
        return "⚠️ Rate limit exceeded."
    except AuthenticationError:
        return "⚠️ Invalid API key."
    except APIError as e:
        return f"⚠️ API error: {e}"
    except Exception as e:
        return f"⚠️ Error: {e}"

# Test it
result = safe_api_call([{"role": "user", "content": "Hello!"}])
print(result)
```

## 8. JSON Output

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Respond only in valid JSON."},
        {"role": "user", "content": """Give info about London:
        {"city": "", "country": "", "population": 0, "famous_for": []}"""}
    ]
)

json_str = response.choices[0].message.content
print("Raw:", json_str)

# Parse it
data = json.loads(json_str)
print(f"\nCity: {data['city']}")
print(f"Country: {data['country']}")
```
