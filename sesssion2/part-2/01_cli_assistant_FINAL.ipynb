## Setup

````python
from openai import OpenAI
import os

# Initialize OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

print("✅ Setup complete!")
````

---
# Part 1: Chat Mode (Conversation with Memory)

## SIMPLE VERSION

````python
# MEMORY

conversation = [
    {"role": "system", "content": "You are a helpful assistant"}
]

while True:
    user_input = input("You: ")
    
    if user_input == "quit":
        break
    
    # Add user message to conversation history
    conversation.append({"role": "user", "content": user_input})
    
    # Send ENTIRE conversation to API
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=conversation  # ← THE KEY!
    )
    
    # Get AI's response and save it too
    ai_message = response.choices[0].message.content
    conversation.append({"role": "assistant", "content": ai_message})
    
    print(f"AI: {ai_message}\n")

print("✅ Simple chat works! Now it remembers the conversation!")
````

### 💡 The Key Concept:

**Memory = A list that keeps growing!**

```python
conversation = [
    {"role": "system", "content": "You are helpful"},
    {"role": "user", "content": "What is AI?"},
    {"role": "assistant", "content": "AI is..."},
    {"role": "user", "content": "Tell me more"},  # ← Can reference previous!
]
```

We send the **ENTIRE list** every time → AI sees full context!

## 📚 COMPLETE VERSION - For Reference

````python
def chat_mode():
    """
    COMPLETE chat mode with all features:
    - Error handling
    - Input validation
    - Nice formatting
    - Multiple exit options
    
    Study this AFTER understanding the simple version above!
    """
    print("=" * 70)
    print("💬 CHAT MODE")
    print("=" * 70)
    print("Start chatting! Type 'quit' to exit.\n")
    
    conversation = [
        {
            "role": "system",
            "content": "You are a helpful, friendly assistant. Provide clear, concise answers."
        }
    ]
    
    while True:
        user_input = input("You: ").strip()
        
        if user_input.lower() in ['quit', 'exit', 'bye', 'q']:
            print("\n✨ Goodbye!\n")
            break
        
        if not user_input:
            print("⚠️  Please enter a message.\n")
            continue
        
        conversation.append({"role": "user", "content": user_input})
        
        try:
            response = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=conversation,
                temperature=0.7,
                max_tokens=300
            )
            
            ai_message = response.choices[0].message.content
            conversation.append({"role": "assistant", "content": ai_message})
            
            print(f"\n🤖 AI: {ai_message}\n")
            
        except Exception as e:
            print(f"\n⚠️  Error: {str(e)}\n")

# Uncomment to run:
# chat_mode()

print("✅ Complete chat mode ready!")
````

# Part 2: Summarize Mode

## 🎯 SIMPLE VERSION - Multi-line Input Helper

````python
# SIMPLE multi-line input function

def get_text():
    """Get multi-line text. Press Enter twice to finish."""
    print("Enter text (press Enter twice when done):\n")
    lines = []
    empty_count = 0
    
    while empty_count < 2:
        line = input()
        if line == "":
            empty_count += 1
        else:
            empty_count = 0
            lines.append(line)
    
    return "\n".join(lines)

print("✅ Multi-line input ready!")
````

## 🎯 SIMPLE VERSION - Summarize Function

````python
# SIMPLE summarize - just 10 lines!

text = get_text()

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Summarize in 2-3 sentences."},
        {"role": "user", "content": f"Summarize: {text}"}
    ]
)

print("\nSummary:", response.choices[0].message.content)
````

## 📚 COMPLETE VERSION - Summarize with Polish

````python
def summarize_mode():
    """
    COMPLETE summarize mode with formatting and validation.
    """
    print("\n" + "=" * 70)
    print("📝 SUMMARIZE MODE")
    print("=" * 70 + "\n")
    
    text = get_text()
    
    if not text.strip():
        print("\n⚠️  No text provided!\n")
        return
    
    print("\n⏳ Generating summary...\n")
    
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Summarize clearly and concisely in 2-3 sentences."},
            {"role": "user", "content": f"Summarize:\n{text}"}
        ],
        temperature=0.3
    )
    
    print("=" * 70)
    print("📊 SUMMARY")
    print("=" * 70)
    print(response.choices[0].message.content)
    print("=" * 70 + "\n")

# Uncomment to test:
# summarize_mode()

print("✅ Summarize mode ready!")
````

# Part 3: Email Writer

## 🎯 SIMPLE VERSION

````python
# SIMPLE email writer - 10 lines!

description = input("What email do you need? ")
tone = input("Tone (formal/friendly/casual): ")

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": f"Write {tone} emails with subject, greeting, body, closing."},
        {"role": "user", "content": f"Write email: {description}"}
    ]
)

print("\nEmail:\n", response.choices[0].message.content)
````

## 📚 COMPLETE VERSION - Email Writer with Polish

````python
def email_mode():
    """
    COMPLETE email writer with validation and formatting.
    """
    print("\n" + "=" * 70)
    print("✉️  EMAIL WRITER MODE")
    print("=" * 70 + "\n")
    
    description = input("Email purpose: ").strip()
    
    if not description:
        print("\n⚠️  Please describe the email!\n")
        return
    
    tone = input("Tone (formal/friendly/casual) [formal]: ").strip()
    if not tone:
        tone = "formal"
    
    print("\n⏳ Writing your email...\n")
    
    system_prompt = f"""You are a professional email writer.
    Write emails in a {tone} tone.
    Include: subject line, greeting, body, and closing."""
    
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": f"Write an email: {description}"}
        ],
        temperature=0.7
    )
    
    print("=" * 70)
    print("✉️  GENERATED EMAIL")
    print("=" * 70)
    print(response.choices[0].message.content)
    print("=" * 70 + "\n")

# Uncomment to test:
# email_mode()

print("✅ Email mode ready!")
````

# Part 4: Complete CLI Assistant with Menu

## 📚 COMPLETE VERSION - Everything Together

````python
def main_menu():
    """
    Complete CLI Assistant with menu system.
    """
    print("\n" + "=" * 70)
    print("🤖 CLI ASSISTANT")
    print("=" * 70)
    print("Your AI-powered command-line helper!\n")
    
    while True:
        print("-" * 70)
        print("Choose a mode:")
        print("  1. 💬 Chat Mode")
        print("  2. 📝 Summarize Mode")
        print("  3. ✉️  Email Writer")
        print("  4. 🚪 Quit")
        print("-" * 70)
        
        choice = input("\nEnter choice (1-4): ").strip()
        
        if choice == '1':
            chat_mode()
        elif choice == '2':
            summarize_mode()
        elif choice == '3':
            email_mode()
        elif choice == '4':
            print("\n" + "=" * 70)
            print("Thank you for using CLI Assistant!")
            print("Goodbye! 👋")
            print("=" * 70 + "\n")
            break
        else:
            print("\n⚠️  Invalid choice. Please enter 1-4.\n")

# Uncomment to run the complete assistant:
main_menu()

print("✅ Complete CLI Assistant ready!")
print("\nUncomment the line above to run!")
````
