```python
students_grade = {}
```

```python
def get_input(prompt):
    return input(prompt).strip().title()
```

```python
def get_grade_input(prompt):
    while True:
        try:
            grade = int(input(prompt))
            if 0 <= grade <= 100:
                return grade
            print("Grade must be between 0 and 100.")
        except ValueError:
            print("Please enter a number.")
```

```python
def doesStudentExist(student_name):
  if student_name in students_grade:
    print(f"{student_name} already exists. Use Update Grade to change it.")
    return True
  return False

def add_grade():
  try:
    student_name = get_input("Enter student name: ")
    if doesStudentExist(student_name):
      return
    grade = get_grade_input("Enter grade: ")
    students_grade[student_name] = grade
    print(f"{student_name}'s grade added successfully!")
  except ValueError:
    print("Invalid input. Please enter a valid student name and grade.")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def show_all_grades():
  if not students_grade:
    print("No grades available.")
  else:
    print("--- All Grades ---")
    for student_name, grade in students_grade.items():
      print(f"{student_name}: {grade}")
```

```python
def show_specific_grade():
  try:
    student_name = get_input("Enter student name: ")
    if student_name in students_grade:
      grade = students_grade[student_name]
      print(f"{student_name}'s grade: {grade}")
    else:
      print(f"Student not found.")
  except ValueError:
    print("Invalid input. Please enter a valid student name.")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def update_grade():
  try:
    student_name = get_input("Enter the student's name to update: ")
    if student_name in students_grade:
      new_grade = get_grade_input("Enter new grade: ")
      students_grade[student_name] = new_grade
      print(f"{student_name}'s grade updated successfully!")
    else:
      print(f"No grade found for {student_name}")
  except ValueError:
    print("Invalid input. Please enter a valid student name.")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def delete_grade():
  try:
    student_name = get_input("Enter the student's name to delete: ")
    if student_name in students_grade:
      del students_grade[student_name]
      print(f"{student_name}'s record deleted successfully!")
    else:
      print(f"Student not found.")
  except ValueError:
    print("Invalid input. Please enter a valid student name.")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def isValidInput(user_choice):
  if user_choice not in ["1", "2", "3", "4", "5", "6"]:
    return False
  return True

def menu():
  print("===== Grade Management System =====")
  print("1. Add Student Grade")
  print("2. Show All Grades")
  print("3. Show Specific Student's Grade")
  print("4. Update Grade")
  print("5. Delete Grade")
  print("6. Exit\n")
  user_choice = input("Choose an option (1–6): ")

  while user_choice != "6":
    if isValidInput(user_choice):
      if(user_choice == "1"):
        add_grade()
      elif(user_choice == "2"):
        show_all_grades()
      elif(user_choice == "3"):
        show_specific_grade()
      elif(user_choice == "4"):
        update_grade()
      elif(user_choice == "5"):
        delete_grade()
    else:
      print("Invalid choice. Please choose a number between 1 and 6.")
    print("\n")
    user_choice = input("Choose an option (1–6): ")

  print("Exiting program. Goodbye!")
```

```python
menu()
```
